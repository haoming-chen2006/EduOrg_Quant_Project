```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=npo["org_year_count"], y=npo["totrev"], alpha=0.5)
plt.xlabel("Years in Operation")
plt.ylabel("Total Revenue")
plt.title("Revenue vs. Years in Operation")
plt.grid(True)
plt.show()
```

```python
bins = [0, 5, 10, 15, 20, 25, 30]
labels = ['0-5', '5-10', '10-15', '15-20', '20-25', '25-30']

# Create the 'year_range' column
npo['year_range'] = pd.cut(npo['org_year_count'], bins=bins, labels=labels, right=False)

distribution = npo.groupby(['year_range', 'Main_Category']).size().unstack(fill_value=0)


total_counts = distribution.sum(axis=0)


top_8_categories = total_counts.sort_values(ascending=False).head(8).index

distribution_top_8 = distribution[top_8_categories]

distribution_percentage_top_8 = distribution_top_8.divide(distribution_top_8.sum(axis=1), axis=0) * 100

plt.figure(figsize=(14, 7))
distribution_percentage_top_5.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Distribution of Top 5 Nonprofit Types Across Different Years of Operation')
plt.xlabel('Years in Operation')
plt.ylabel('Percentage of NGOs')
plt.legend(title='Nonprofit Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

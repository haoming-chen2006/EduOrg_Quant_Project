```python
def get_quant_cols(df):
    return df[['totrev', 'total_income', 'total_assets', 'org_year_count', 'org_fiscal_period', 'org_year_first', 'org_year_last']]
```

```python
org = np.array(["Scholarship","Single Support Fundraising","Fundraising","Education Services","Student Services","Advocacy","Research"])
orgs = npo[npo.Main_Category.isin(org)]
df = get_quant_cols(orgs[orgs.Main_Category == "Scholarship"])
single_support_fundraising = get_quant_cols(orgs[orgs.Main_Category == "Single Support Fundraising"])
fundraising = get_quant_cols(orgs[orgs.Main_Category == "Fundraising"])
education_services = get_quant_cols(orgs[orgs.Main_Category == "Education Services"])
student_services = get_quant_cols(orgs[orgs.Main_Category == "Student Services"])
advocacy = get_quant_cols(orgs[orgs.Main_Category == "Advocacy"])
research = get_quant_cols(orgs[orgs.Main_Category == "Research"])
scholarship = get_quant_cols(orgs[orgs.Main_Category == "Scholarship"])
```

```python
"""Get new cols through feature engineering"""
def engineer_features(df):
    df["total_income_per_asset"] = df["total_income"] / df["total_assets"]
    df["total_income_per_year"] = df["total_income"] / df["org_year_count"]
    df["total_income/totrev"] = df["total_income"] / df["totrev"]
    df["total_income/total_asstes"] = df["total_income"] / df["total_assets"]
    return df

# Replace positive and negative infinity values with 1
def replace_infinity_with_one(df):
    return df.replace([np.inf, -np.inf], 1)

# Apply feature engineering and replace infinity values with 1
single_support_fundraising = replace_infinity_with_one(engineer_features(single_support_fundraising).fillna(1))
fundraising = replace_infinity_with_one(engineer_features(fundraising).fillna(1))
education_services = replace_infinity_with_one(engineer_features(education_services).fillna(1))
student_services = replace_infinity_with_one(engineer_features(student_services).fillna(1))
advocacy = replace_infinity_with_one(engineer_features(advocacy).fillna(1))
research = replace_infinity_with_one(engineer_features(research).fillna(1))
scholarship = replace_infinity_with_one(engineer_features(scholarship).fillna(1))


```

```python
import numpy as np
from sklearn.ensemble import IsolationForest
# Separate the non-numeric columns (e.g., organization names)
"""Count nas in research"""
i = 0
def get_outliers(df):
    model = IsolationForest(contamination=0.03, n_estimators=100, random_state=42)
    model.fit(df)
    df["outlier"] = model.predict(df)
    outliers = df[df['outlier'] == -1]
    
    # Extract the middle value (position 1) of the index for each outlier
    outlier_names = [index[1] for index in outliers.index]
    print(outliers.shape[0])
    
    return outliers

# Example usage
scholarship_outlier = get_outliers(scholarship)
single_support_fundraising_outlier = get_outliers(single_support_fundraising)
fundraising_outlier = get_outliers(fundraising)
education_services_outlier = get_outliers(education_services)
student_services_outlier = get_outliers(student_services)
advocacy_outlier = get_outliers(advocacy)
research_outlier = get_outliers(research)


        
```
